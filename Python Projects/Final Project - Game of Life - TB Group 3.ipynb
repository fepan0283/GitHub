```python
import random
import time

# This is our dictionary of fields. The game loop accesses the different action fields (e.g.q5) when the player reaches a 
# specific field (e.g. 7). 
field_dic = {
    1: q1,
    2: q2,
    3: q3,
    4: q4,
    5: q5,
    6: q6,
    7: q7,
    8: q8,
    9: q9,
    10: q10,
    11: q11,
    12: q12,
    13: q5,
    14: q14,
    15: q15,
    16: q16,
    17: q17
}
# At the beginning of the game, the players can decide how many people are playing. Initially, all 4 instances of the player 
# class are created but only the amount of desired players are part of the following game loop
class player_creation:
    def __init__(self, number, name, field, balance, university, job, salary, children, turn):
        self.number = number
        self.name = name
        self.field = field
        self.balance = balance
        self.university = university
        self.job = job
        self.salary = salary
        self.children = children
        self.turn = turn
player_1 = player_creation(1, "Player", 0, 0, "tbd", "tbd", 0, 0, 0)
player_2 = player_creation(2, "Player", 0, 0, "tbd", "tbd", 0, 0, 0)
player_3 = player_creation(3, "Player", 0, 0, "tbd", "tbd", 0, 0, 0)
player_4 = player_creation(4, "Player", 0, 0, "tbd", "tbd", 0, 0, 0)

list_of_players = [player_1, player_2, player_3, player_4]


# At the beginning of the game, players enter their name, depending on the number of players they choose to play with
def naming_players(number_of_players):
    if number_of_players == 1:
        name_1 = input("Player 1, what is your name?")
        print("")
        player_1.name = name_1
    elif number_of_players == 2:
        player_1.name = input("Player 1, what is your name?")
        player_2.name = input("Player 2, what is your name?")
        print("")
    elif number_of_players == 3:
        player_1.name = input("Player 1, what is your name?")
        player_2.name = input("Player 2, what is your name?")
        player_3.name = input("Player 3, what is your name?")
        print("")
    elif number_of_players == 4:
        player_1.name = input("Player 1, what is your name?")
        player_2.name = input("Player 2, what is your name?")
        player_3.name = input("Player 3, what is your name?")
        player_4.name = input("Player 4, what is your name?")
        print("")
    

# Throwing the dice. Every player is assigned a random number that she advances on the board. 
def throw_dice(player):
    print("Its your turn,",player.name)
    random_number = random.randrange(1,4)
    player.field += random_number
    player.turn += 1
    print("You threw a", random_number, "!")

# This accesses the actual field action dictionary defined above
def field_action(player):
    field_dic[player.field](player)
    
# Every two rounds the players get their salary that is added to their balance
def pay_day(player):
    if (player.turn % 2) == 0 and player.job != "tbd":
        player.balance += player.salary
        print("It is Payday! Your balance will be updated by the amount of your salary:", player.salary)

# The start of the game comprises the selection of players and the naming of players
number_of_players = int(input("Hello players. Welcome to the Game of Life! How many people are playing? You can choose between 1-4"))
naming_players(number_of_players)
print("Great. Lets get started. The rules are easy. You throw the dice and land on a new field. An event happens and you will either profit of lose from it. Sometimes you will have to make decision as well. Good luck!")
print("")
time.sleep(2)

# At the end of the game every player receives their final balances and information about their achievements in life.
# The winner, the player with the highest balance, is then determined
def end_result():
    dic_of_balances = {}
    for player in list_of_players:
        if player.name != "Player":
            dic_of_balances[player.name] = player.balance
            print("Hey", player.name, "! Your final balance is", player.balance)
            print("As a", player.job,", you earned", player.salary, "EUR a year and had", player.children, "children.")
            print("")
    print("And the winner of the game is", max(dic_of_balances, key=dic_of_balances.get).upper(), "! Congratulations!")
    

# game loop. The while loop plays through every player until they reach the final field. Along the way, they make decisions
# and gain or lose money. The final loop ends when all players reach the final field and the winner is announced
end_game_counter = 0
while end_game_counter == 0:
    for player in list_of_players:
        if player.turn == 1:
            choose_education(player)
            player.turn += 1
            print("")
            time.sleep(4)
        elif player.turn == 2:
            choose_job(player)
            player.turn += 1
            print("")
            time.sleep(4)
        elif int(number_of_players) >= player.number and player.field <= 17:
            throw_dice(player)
            if player.field <= 17:
                field_action(player)
                pay_day(player)
            else:
                print("Congratulations, you managed to finished the game")
            print("")
            time.sleep(4)
        elif int(number_of_players) >= player.number and player.field > 17:
            print("Congratulations to everybody, this is the end of the game")
            end_result()
            end_game_counter += 1
            break

            
```
